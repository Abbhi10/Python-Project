```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ICC Mens T20 Worldcup.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Univariate analysis - Distribution of first innings scores
plt.figure(figsize=(10, 5))
sns.histplot(df['First Innings Score'], bins=20, kde=True, edgecolor='black', alpha=0.7)
plt.title("Distribution of First Innings Scores")
plt.xlabel("First Innings Score")
plt.ylabel("Density")
plt.show()

# Univariate analysis - Toss decision
plt.figure(figsize=(6, 4))
sns.countplot(x=df['Toss Winning'])
plt.xticks(rotation=90)
plt.title("Toss Winners Distribution")
plt.xlabel("Teams")
plt.ylabel("Count")
plt.show()
```

```python
# Bivariate analysis - Relationship between first innings score and match winner
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Winners'], y=df['First Innings Score'])
plt.xticks(rotation=90)
plt.title("First Innings Score vs Match Winner")
plt.xlabel("Match Winner")
plt.ylabel("First Innings Score")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Encoding categorical variables
encoder = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
df_encoded = df.copy()

for col in categorical_cols:
    df_encoded[col] = encoder.fit_transform(df[col].astype(str))

# Correlation analysis - Heatmap for numerical features
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Outlier detection - Winning margins
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Winning Margin'])
plt.title("Outlier Detection in Winning Margin")
plt.xlabel("Winning Margin (Runs/Wickets)")
plt.show()
```

```python
# Key insights
print("\nKey Findings:")
print("1️ Most first innings scores range between 120-180 runs.")
print("2️ Teams winning the toss often choose to chase.")
print("3️ Higher first innings scores generally lead to match wins.")
print("4️ A few matches had exceptionally high winning margins (above 100 runs or 9-10 wickets).")
```
